```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Week1_Day1_Foundation") \
    .getOrCreate()
```

```python
data = [
    (1, "Alice", 25, 5000),
    (2, "Bob", 30, 7000),
    (3, "Charlie", 22, 3000),
    (4, "Diana", 35, 9000),
    (5, "Eve", 28, 6000)
]

columns = ["id", "name", "age", "salary"]

df = spark.createDataFrame(data, columns)
```

```python
# Basic CSV read — Spark infers types (often gets them wrong)
df = spark.read.csv("employees.csv", header=True, inferSchema=True)

# Check what you loaded
df.show(10)          # First 10 rows, formatted as table
df.printSchema()     # Column names + data types
df.describe().show() # Count, mean, stddev, min, max for numeric cols
df.count()           # Total row count
```

```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DoubleType

schema = StructType([
    StructField("emp_id", IntegerType(), nullable=False),
    StructField("name", StringType(), nullable=False),
    StructField("department", StringType(), nullable=True),
    StructField("salary", DoubleType(), nullable=True),
    StructField("hire_date", StringType(), nullable=True)
])

df = spark.read.csv("employees.csv", header=True, schema=schema)
df.printSchema()
```
